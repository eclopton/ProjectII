> If you've taken a trip to the nearest movie theater, chances are you got out of the box office wondering whether or not you paid for a ticket stub or your own personal highway robbery. With that question in mind, it's time to look at a number of statistics released each year by the Motion Picture Association of America as well as the National Association of Theatre Owners to see if tickets really do seem to getting more expensive.

```r
library(ggplot2)
library(dplyr)
library(reshape2)
```

```r
totals <- read.csv("Desktop/Movie Project/totals.csv")
```

```r
tickets <- read.csv("Desktop/Movie Project/ticketprices.csv")
```

```r
ggplot(tickets, aes(x=Year, y=Price)) + geom_line() +
labs(x="Year", y="U.S. movie ticket prices ($)", title="Average U.S. movie ticket prices ", subtitle="American ticket stubs have increased steadily since 1997", caption="Source: National Association of Theatre Owners | Graphic By: Ellis Clopton")
```

> By looking at this graph, it appears that movie ticket prices have been increasing since 1997 from the average U.S. ticket costing less than five dollars to costing around nine dollars last year. This timeframe was chosen to reduce the impact of inflation on the graph. With the increasing ticket prices, how has this impacted box office performances?

```r
ggplot(totals, aes(x=Year, y=Box.Office)) + geom_line() +
labs(x="Year", y="Domestic Box Office Gross in billions", title="Box office growth since 1997", subtitle="Total box office earnings have grown in the United States
and Canada", caption="Source: Motion Picture Association of America | Graphic By: Ellis Clopton")
```

> Just eyeballing this line chart, it would appear that box office earnings have been growing steadily since the beginning of the 21st century, with a few dips here and there. However, looking closer at around 2010 we start to see those dips becoming more frequent, most notable in the last year. These dips also seem to correlate to when ticket prices jump in price.

> So how come the ticket prices are rising still, even in years where attendance drops? The answer may be in not how many people show up, but what kind of movie theatre they end up going to. For this, let's look at the number of cinema sites in America versus the number of screens.

```r
screens <- read.csv("Desktop/Movie Project/screens.csv")
```

```r
ggplot(screens, aes(Year)) + 
  geom_line(aes(y = Sites, colour = "Sites")) + 
  geom_line(aes(y = Total, colour = "Screens")) + labs(title= "American cinema screens versus American cinema locations", subtitle="As cinema cites decrease, the number of cinema screens increase", caption="National Association of Theatre Owners | Graphic By: Ellis Clopton")
```

> Interesting, right? Even though the amount of theatres in the country have been steadily dropping over the past decade, the number of screens is only increasing, with a sharp increase right before 2000. This is possible because, like any widespread profitable business venture, the growth is in corporate cinema chains while smaller theatres close their doors.

> As every moviegoer knows, these larger multiplex cinemas offer a wide array of services that focus on the customer's experience. This will obviously lead to higher ticket prices, but you may be wondering where all this money goes and what is causing your favorite chain to charge your family the price of a tank of gas for a night out. Thankfully, America's top-earning circuit, AMC theaters, has the numbers.

```r
costs <- read.csv("Desktop/Movie Project/costs.csv")
```

```r
revenue <- read.csv("Desktop/Movie Project/revenues.csv")
```

```r
ggplot(revenue, aes(1, Totals, fill=Revenues)) + geom_bar(stat="identity") + labs(x="AMC Revenues", y="Totals in millions ($)", title="AMC Revenues at the end of quarter 4, 2017", caption="AMC Entertainment Holdings, Inc. | Graphic By: Ellis Clopton")
```

> Surprise, surprise: a big chunk of AMC's revenue comes from ticket stubs. However, it should be worth remembering that cinemas pay a chunk of that revenue, generally around 60 percent, to movie studios in order to lease their films. Let's factor that expense along with others in another stacked bar chart.

```r
ggplot(costs, aes(1, Total, fill=Cost)) + geom_bar(stat="identity") + labs(x= "Costs of running a cinema", y= "Totals in millions ($)", title = "AMC costs and expenses at the end of quarter 4, 2017", caption="AMC Entertainment Holdings, Inc. | Graphic By: Ellis Clopton")
```

> Ouch, on top of the general operating costs, paying out those film exhibiton leasing fees take a large chunk out of your revenue. So where do movie theaters make most of their money? Food and beverages. Many of these chains, including AMC and, more recently, Alamo Drafthouse have begun investing more of their expenses into turning their cinemas into somewhat of a dining experience. With the increasing utilization of dining options and installing recliners jacking up the prices of tickets, one has to wonder why these larger chains are leaning so hard into changing how we see new movies.
